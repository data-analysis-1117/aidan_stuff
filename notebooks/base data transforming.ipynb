# Inspection of data

### notes on what transformations are done

 - columns with only nan deleted
 - columns with no variance deleted
 - normalized values

```python
# imports
import pandas as pd
```

```python
training = pd.read_csv("../data/original/training.csv", delimiter = ";")
holdout = pd.read_csv("../data/original/holdout.csv", delimiter = ";")
```

```python
training.head(5)
```

```python
# target shows whether the company was created for fraudulent purposes or not
# converting target variables to just True and False
target_new = list()

for idx, row in training.iterrows():
    target = row['target']
    if target == "target_FALSE":
        target_new.append(False)
    elif target == "target_TRUE":
        target_new.append(True)

training.drop('target', axis=1, inplace=True)
training['target'] = target_new

training['target']
```

```python
# print all columns with 3 example samples
for (columnName, columnData) in training.iteritems():
    print(columnName)
    print(columnData.values[:3])
    print(type(str(columnData.values[0])))
    print()
```

```python
# printing nan percentages
row_count = training.shape[0]
for (columnName, columnData) in training.iteritems():
    print(columnName)
    
    nan_count = 0
    for point in columnData:
        point = str(point)
        if point == "nan":
            nan_count+=1
    
    nan_percent = nan_count/row_count
    print(nan_percent)
    print()
```

```python
# deleting all columns with a nan amount of 100%
drop_list = list()

# print all columns with 3 example samples
row_count = training.shape[0]
for (columnName, columnData) in training.iteritems():
    
    nan_count = 0
    for point in columnData:
        if str(point) == 'nan':
            nan_count += 1
            
    nan_percent = nan_count / row_count
    
    if nan_percent > 0.3:
        drop_list.append(columnName)

print(drop_list)
```

```python
# dropping them
for columnName in drop_list:
    training.drop(columnName, axis=1, inplace=True)
    
training.head(5)
```

```python
# deleting columns with no variance
del_list = list()
for (columnName, columnData) in training.iteritems():
    unique_values = set()
    
    for point in columnData:
        point = str(point)
        unique_values.add(point)
    
    unique_values = list(unique_values)
    
    if len(unique_values) == 1:
        del_list.append(columnName)
    
print(del_list)

for columnName in del_list:
    training.drop(columnName, axis=1, inplace=True)
```

```python
# printing distribution of fraudulent to non fraudulent rows
print(training['target'].value_counts())
```
